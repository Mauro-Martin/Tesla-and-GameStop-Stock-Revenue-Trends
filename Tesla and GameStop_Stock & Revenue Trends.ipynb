# <h1>Question 1: Mining Tesla stock data using yfinance<h1>

The `yfinance` library is used to download historical stock data for Tesla, including closing prices, volume, and other variables. The date index is reset as a column for easier manipulation, and the first five rows of the dataset are displayed.

```python
import yfinance as yf

# Download Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Reset the index to make 'Date' a column instead of an index
tesla_data.reset_index(inplace=True)

# Save and display the first five rows
tesla_data.head()
```

# <h1>Question 2: Extract Tesla Revenue Data Using Webscraping<h1>

Using `requests` and `BeautifulSoup`, an HTTP request is made to extract Tesla's quarterly revenue from an HTML table. The data is processed by cleaning the revenue values and storing them in a DataFrame. Revenue values are converted to numeric format, and rows with missing data are dropped. The last five rows of the DataFrame are displayed.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for Tesla's quarterly revenue
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'

# Perform the HTTP request
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all tables on the page
tables = soup.find_all('table')

# Select the correct table
relevant_table = tables[1]

# Create an empty DataFrame with the desired columns
tesla_revenue = pd.DataFrame(columns=['Date', 'Revenue'])

# Locate the table body
tbody = relevant_table.find('tbody')

# Process each row
rows = tbody.find_all('tr') if tbody else relevant_table.find_all('tr')

for row in rows:
    columns = row.find_all('td')
    if len(columns) == 2:
        date = columns[0].text.strip()
        revenue = columns[1].text.strip().replace('$', '').replace(',', '')
        if revenue:
            tesla_revenue = pd.concat([tesla_revenue, pd.DataFrame([[date, revenue]], columns=['Date', 'Revenue'])], ignore_index=True)

# Convert 'Revenue' column to numeric, handle conversion errors
tesla_revenue['Revenue'] = pd.to_numeric(tesla_revenue['Revenue'], errors='coerce')

# Drop rows with NaN values in the 'Revenue' column
tesla_revenue.dropna(subset=['Revenue'], inplace=True)

# Display the last 5 rows of the DataFrame
tesla_revenue.tail()
```

# <h1>Question 4: Extraction of GameStop Revenue Data using Web Scraping<h1>

Similar to Tesla, GameStop's quarterly revenue data is extracted from an HTML table using `requests` and `BeautifulSoup`. The data is cleaned and stored in a DataFrame, converting revenues to numeric format and dropping rows with missing values. The last five rows of the DataFrame are displayed.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for GameStop's quarterly revenue
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html'

# Perform the HTTP request
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all tables on the page
tables = soup.find_all('table')

# Select the correct table
relevant_table = tables[1]

# Create an empty DataFrame with the desired columns
gamestop_revenue = pd.DataFrame(columns=['Date', 'Revenue'])

# Locate the table body
tbody = relevant_table.find('tbody')

# Process each row
rows = tbody.find_all('tr') if tbody else relevant_table.find_all('tr')

for row in rows:
    columns = row.find_all('td')
    if len(columns) == 2:
        date = columns[0].text.strip()
        revenue = columns[1].text.strip().replace('$', '').replace(',', '')
        if revenue:
            gamestop_revenue = pd.concat([gamestop_revenue, pd.DataFrame([[date, revenue]], columns=['Date', 'Revenue'])], ignore_index=True)

# Convert 'Revenue' column to numeric, handle conversion errors
gamestop_revenue['Revenue'] = pd.to_numeric(gamestop_revenue['Revenue'], errors='coerce')

# Drop rows with NaN values in the 'Revenue' column
gamestop_revenue.dropna(subset=['Revenue'], inplace=True)

# Display the last 5 rows of the DataFrame
gamestop_revenue.tail()
```

# <h1>Definition of the make_graph function<h1>

The `make_graph` function generates interactive plots showing the evolution of stock prices and historical revenues of a company. Using `plotly`, two subplots are created, one for the share prices and one for revenues, allowing for axis customization and titles.

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing=.3)
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data.Date), y=stock_data.Close.astype("float"), name="Share Price"), row=1, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data.Date), y=revenue_data.Revenue.astype("float"), name="Revenue"), row=2, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    fig.update_layout(showlegend=False, height=900, title=stock, xaxis_rangeslider_visible=True)
    fig.show()
```

# <h1>Question 5: Dashboard of Tesla Stock and Revenue<h1>

This code generates a graph showing the historical evolution of Tesla's stock prices and revenue over time.

```python
make_graph(tesla_data, tesla_revenue, 'Evolution of Tesla Shares and Income')
```

# <h1>Question 6: GameStop Stock and Revenue Dashboard<h1>

This code generates a graph displaying the financial trends of GameStops stock prices and revenues.

```python
make_graph(gme_data, gamestop_revenue, 'GameStop Financial Trend: Shares and Income')
```
